## Aplicamos Rolling Forecasting para evaluación del modelo

```python
def crossvalidation():        
    predictions_rolling = pd.DataFrame({})
    test = test_data.copy()
    test = test.reset_index()
    test = test.drop(columns=['y'])    
    
    for end_date in test_data.index:
        print(end_date)
        train_data = Periodo1[:end_date - timedelta(hours=1)]  #Entrenamos el modelo cada que avanza 1 hora
        train_data = train_data.reset_index()
        m = Prophet()
        m.add_seasonality(name='monthly', period=30.5, fourier_order=5)
#         m.add_seasonality(name='Lunes24AM_season', period=121, fourier_order=5, condition_name='Lunes24AM')
#         m.add_seasonality(name='Lunes1AM_season', period=121, fourier_order=5, condition_name='Lunes1AM')
#         m.add_seasonality(name='Lunes2AM_season', period=121, fourier_order=5, condition_name='Lunes2AM')
#         m.add_seasonality(name='Lunes3AM_season', period=121, fourier_order=5, condition_name='Lunes3AM')
#         m.add_seasonality(name='Lunes4AM_season', period=121, fourier_order=5, condition_name='Lunes4AM')
#         m.add_seasonality(name='Lunes5AM_season', period=121, fourier_order=5, condition_name='Lunes5AM')
#         m.add_seasonality(name='Lunes6AM_season', period=121, fourier_order=5, condition_name='Lunes6AM')
#         m.add_seasonality(name='Lunes7AM_season', period=121, fourier_order=5, condition_name='Lunes7AM')
        m.add_seasonality(name='Lunes8AM_season', period=121, fourier_order=5, condition_name='Lunes8AM')
#         m.add_seasonality(name='Lunes9AM_season', period=121, fourier_order=5, condition_name='Lunes9AM')
#         m.add_seasonality(name='Lunes10AM_season', period=121, fourier_order=5, condition_name='Lunes10AM')
#         m.add_seasonality(name='Lunes11AM_season', period=121, fourier_order=5, condition_name='Lunes11AM')
#         m.add_seasonality(name='Lunes12PM_season', period=121, fourier_order=5, condition_name='Lunes12PM')
#         m.add_seasonality(name='Lunes13PM_season', period=121, fourier_order=5, condition_name='Lunes13PM')
        m.add_seasonality(name='Lunes14PM_season', period=121, fourier_order=5, condition_name='Lunes14PM')        
#         m.add_seasonality(name='Lunes15PM_season', period=121, fourier_order=5, condition_name='Lunes15PM')        
        m.add_seasonality(name='Lunes16PM_season', period=121, fourier_order=5, condition_name='Lunes16PM')
#         m.add_seasonality(name='Lunes17PM_season', period=121, fourier_order=5, condition_name='Lunes17PM')
#         m.add_seasonality(name='Lunes18PM_season', period=121, fourier_order=5, condition_name='Lunes18PM')
#         m.add_seasonality(name='Lunes19PM_season', period=121, fourier_order=5, condition_name='Lunes19PM')
#         m.add_seasonality(name='Lunes20PM_season', period=121, fourier_order=5, condition_name='Lunes20PM')
#         m.add_seasonality(name='Lunes21PM_season', period=121, fourier_order=5, condition_name='Lunes21PM')
#         m.add_seasonality(name='Lunes22PM_season', period=121, fourier_order=5, condition_name='Lunes22PM')
#         m.add_seasonality(name='Lunes23PM_season', period=121, fourier_order=5, condition_name='Lunes23PM')

        m.fit(train_data);                     
        
        future = test.loc[test.loc[:,'ds'] == end_date]        
        forecast = m.predict(future)        
        output=forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']]
        yhat = output[['ds','yhat']]
        predictions_rolling = predictions_rolling.append(yhat)       
        
        
        
    predictions_rolling = predictions_rolling.set_index('ds')
    fig = m.plot_components(forecast)
        
    return predictions_rolling, fig
```

```python
predictions_rolling, fig = crossvalidation()
```
